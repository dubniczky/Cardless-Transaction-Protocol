```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

method_names = ['Crystals Dilithium 2', 'Crystals Dilithium 3', 'Crystals Dilithium 5', 'Falcon 512', 'Falcon 1024',
    'Sphincs SHA256-128f Simple', 'Sphincs SHA256-192f Simple', 'Sphincs SHA256-256f Simple']
df = pd.DataFrame(np.array([
    [1, 1312, 2528, 2420, 116511, 342726, 112506],
    [3, 1952, 4000, 3293, 191331, 534254, 180350],
    [5, 2592, 4864, 4595, 307765, 610807, 417971],
    [1, 897, 1281, 690, 24656358, 1085984, 183949],
    [5, 1793, 2305, 1330, 74741342, 2204927, 359553],
    [1, 32, 64, 17088, 3088404, 72191077, 8962488],
    [3, 48, 96, 35664, 3920103, 119085653, 12269960],
    [5, 64, 128, 49856, 10771651, 220637782, 12168947]
]), columns=['Security level', 'Public key size (bytes)', 'Private key size (bytes)', 'Signature size (bytes)', 'Keygen (CPU cycles)', 'Sign (CPU cycles)', 'Verify (CPU cycles)'])
X_axis = np.arange(len(method_names))

df
```

```python
plt.bar(X_axis - 0.3, df['Public key size (bytes)'], 0.3, label = 'Public key size')
plt.bar(X_axis, df['Private key size (bytes)'], 0.3, label = 'Private key size')
plt.bar(X_axis + 0.3, df['Signature size (bytes)'], 0.3, label = 'Signature size')

plt.xticks(X_axis, method_names, rotation = 90)
plt.yscale('log')
plt.xlabel('Methods')
plt.ylabel('Bytes')
plt.title('Key and signature sizes')
plt.legend()
plt.show()
```

```python
plt.bar(X_axis - 0.3, df['Keygen (CPU cycles)'], 0.3, label = 'Keygen')
plt.bar(X_axis, df['Sign (CPU cycles)'], 0.3, label = 'Sign')
plt.bar(X_axis + 0.3, df['Verify (CPU cycles)'], 0.3, label = 'Verify')

plt.xticks(X_axis, method_names, rotation = 90)
plt.yscale('log')
plt.xlabel('Methods')
plt.ylabel('CPU cycles')
plt.title('Performance')
plt.legend()
plt.show()
```
